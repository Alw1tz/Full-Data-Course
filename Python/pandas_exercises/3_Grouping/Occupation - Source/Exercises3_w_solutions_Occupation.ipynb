### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean(numeric_only = True)['age']
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# 1st form 
# explicacion en español: gender_to_numeric() para convertir los valores de género en la columna 'gender' de un dataframe de usuarios (users) de la cadena 'M' o 'F' a un valor numérico (1 para 'M' y 0 para 'F') y crea una nueva columna llamada 'gender_n' con los valores numéricos.
# Luego, se agrupa el dataframe por ocupación (users.groupby('occupation')) y se calcula la suma de la columna 'gender_n' (que representa la cantidad de hombres por ocupación) dividida por el número total de registros por ocupación (users.occupation.value_counts()). Multiplicando este resultado por 100 se obtiene el porcentaje de hombres por ocupación.
# Finalmente, los resultados se ordenan de mayor a menor utilizando la función sort_values() con el argumento ascending = False.

def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

# apply the function to the gender column and create a new column
users['gender_n'] = users['gender'].apply(gender_to_numeric)


a = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100 

# sort to the most male 
a.sort_values(ascending = False)
```

```python
# 2nd form
conteo = users.groupby(['occupation', 'gender']).count()

men_by_occupation = conteo.loc[(slice(None), 'M'), :]

men_ratio = round((men_by_occupation / conteo.groupby(level=0).sum() * 100), 4)

men_ratio = men_ratio.sort_values(by='age', ascending=False)
men_ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(minimum_age=('min'), maximum_age=('max'))
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].mean(numeric_only = True)
```

### Step 8.  For each occupation present the percentage of women and men

```python
# 1st form
# create a data frame and apply count to gender
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

# create a DataFrame and apply count for each occupation
occup_count = users.groupby(['occupation']).agg('count')

# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100

# present all rows from the 'gender column'
occup_gender.loc[: , 'gender']
```

```python
# explicacion en español:
# Agrupar los datos por ocupación y género, y contar el número de hombres y mujeres en cada grupo: es decir, te contamos cuantos hay en total en M y cuantos en total en F
gender_counts = users.groupby(['occupation', 'gender'])['age'].count()

# # Calcular el número total de personas en cada ocupación: es decir, el total de M y F en cada ocupación
total_counts = users.groupby('occupation')['age'].count()

# # Calcular el porcentaje de hombres y mujeres en cada ocupación
percentages = gender_counts.groupby('occupation', group_keys=True).apply(lambda x: 100 * x / float(total_counts[x.name]))
percentages
```
