### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chipo
```

### Step 4. How many products cost more than $10.00?

```python
# Functional programming
# 1st : item_price (object) -> item_price (float)
plus10 = chipo['item_price'].astype(float) 

# cleaned data
chipo['item_price'] = plus10

def filter(df, column, wdata):
    mascara = df[column] > 10
    valores_mayores = df[mascara]
    return valores_mayores

filter(chipo, 'item_price', 'item_name')
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
def filter_2_columns(df, column, wdata, condition):
    mascara = (df[column] > 10) & (df[condition] == 1)
    valores_mayores = df[mascara][[wdata, column]]
    return valores_mayores

filter_2_columns(chipo, 'item_price', 'item_name', 'quantity')
```
